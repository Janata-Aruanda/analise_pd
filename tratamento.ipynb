```python
import pandas as pd 
tabela1 = pd.read_csv(r'churn-bigml-80.csv')
tabela2 = pd.read_csv(r'churn-bigml-20.csv')
display(tabela1)
display(tabela2)

```

# Concat

```python
# juntando tabelas com o metodo concat/
tabela_teste = pd.concat([tabela1 , tabela2], ignore_index=True)
display(tabela_teste)
```

```python
#Mostra a Informação da tabela
#print(tabela.info())
#Deletar colunas Vazias 
tabela_teste = tabela_teste.dropna(how='all', axis=1)

#Deletar linhas vazias 
tabela_teste = tabela_teste.dropna(how='any', axis=0)

display(tabela_teste.info())
```

```python
# Mostrando os cancelamentos ( Churn = cancelamentos)
print(tabela_teste['Churn'].value_counts())

# Mostrando os cancelamentos ( Churn = cancelamentos)     (Formatação %)
print(tabela_teste['Churn'].value_counts(normalize=True).map("{:.1%}".format))
```
